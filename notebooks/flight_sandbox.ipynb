```python
delay = flights_df.groupby(['FLIGHT_NUMBER'], as_index=True)['ELAPSED_TIME'].sum().reset_index()
```

```python
delay_sub = delay.sample(n=1000)
```

```python
from bokeh.plotting import figure, output_file, show
import bokeh.plotting as bk
bk.output_notebook()

x=delay_sub['FLIGHT_NUMBER']
y=delay_sub['ELAPSED_TIME']

p = figure()
p.circle(x,y,fill_color="red",size=2)
#p.line(x,y,line_width=2)

#output_file("scatter.html")

bk.show(p)
```
